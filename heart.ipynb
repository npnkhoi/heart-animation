```python

def get_z(x: float, y: float): 
    k = x*x+9/80*y*y
    
    a = 1
    b = -(k ** (1.0/3))
    c = x*x + 9/4*y*y - 1
    
    delta = b**2 - 4*a*c
    if delta < 0:
        return []
    elif delta == 0:
        return [-b/2.0/a]
    else:
        delta_root = math.sqrt(delta)
        return [(-b+delta_root)/2/a, (-b-delta_root)/2/a]

    
```

```python
X, Y, Z = [], [], []
for x in np.arange(-2.0, 2.0, 0.05):
    for y in np.arange(-2.0, 2.0, 0.05):
        z_list = get_z(x, y)
        for z in z_list:
            X.append(x)
            Y.append(y)
            Z.append(z)

fig = plt.figure(dpi=120)
ax = plt.axes(projection='3d')
ax.scatter3D(X, Y, Z, c=Z, cmap='Greens');
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

for angle in range(0, 360):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(.001)
```
